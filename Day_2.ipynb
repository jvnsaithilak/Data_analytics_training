```python
import statistics
import matplotlib.pyplot as plt


member_names=["ananya","pooja","vivek","suri","sameer"]
ages=[22,25,24,23,26]
workout_hours=[2,1.5,2.5,1,3]
calories_burned=[500,300,450,600,700]
print("---------------Fitness center Dataset---------------")
for i in range(len(member_names)):
  print(member_names[i],ages[i],workout_hours[i],calories_burned[i])

average_workout=sum(workout_hours)/len(workout_hours)
print("Average workout hours:",average_workout)

highest_calories=max(calories_burned)
print("highest calories burned:",highest_calories)

lowest_calories=min(calories_burned)
print("lowest calories burned:",lowest_calories)

lowest_workout=min(workout_hours)
print("lowest workout hours:",lowest_workout)

median_calories=statistics.median(calories_burned)
print("Median calories burned:",median_calories)

std_dev=statistics.stdev(calories_burned)
print("Standard deviation:",std_dev)

index=calories_burned.index(highest_calories)
print("Most active member:",member_names[index])

index=calories_burned.index(lowest_calories)
print("inactive member:",member_names[index])
print("-------------------charts--------------------")

plt.pie(workout_hours,labels=member_names)
plt.title("Fitness Analysis")
plt.show()

plt.bar(member_names,calories_burned)
plt.title("Fitness Analysis")
plt.xlabel("Member names")
plt.ylabel("Calories burned")
plt.show()

plt.plot(member_names, workout_hours)
plt.title("Fitness Analysis")
plt.xlabel("Member names")
plt.ylabel("Workout hours")
plt.show()

plt.scatter(member_names, calories_burned)
plt.title("Fitness Analysis")
plt.xlabel("Member names")
plt.ylabel("Calories burned")
plt.show()
```
