```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

marks=[90,95,90,70,85,75]
hours=[7,9,8,7,9,7]

print('mean:',np.mean(marks))
print('median:',np.median(marks))
print('mode:',pd.Series(marks).mode())
print('variance:',np.var(marks))
print('standard deviation:',np.std(marks))
print('correlation:',np.corrcoef(marks,hours))

plt.figure(figsize=(5,5))
plt.hist(marks)
plt.xlabel('marks')
plt.ylabel('hours')
plt.title('marks vs hours')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel('example.xlsx')
print("-------customer data-----------")
print(data)
df=pd.DataFrame(data)
print("--------total abandonment--------")
abandoned= df[df['purchase_completed'] == 0]
for name in abandoned['cust_name']:
    print(name)
print("-------internet ----------")
internet= df[df['internet_speed'] <=2]
for name in internet['cust_name']:
    print(name)
print("----------payment--------")
payment=df[df['payment_failure'] <= 1]
for name in payment['cust_name']:
    print(name)
print("-----------Delivery--------")
delivery=df[df['delivery_charges'] >= 50]
for name in delivery['cust_name']:
    print(name)
print("--------root cause-------")
for i in range(len(df['cust_name'])):
  if(df['purchase_completed'][i]==0 and df['internet_speed'][i]<=2 and df['payment_failure'][i]<=1 and df['delivery_charges'][i]>=50):
    print(df['cust_name'][i])
```
