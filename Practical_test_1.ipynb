```python
#food Delivery Analysis
from numpy import average
import pandas as pd
import matplotlib.pyplot as plt

Order_Time=[6,7,8,8,9,10]
Delivery_Time=[7.15,8.25,9.30,9.35,10.20,11.10]
Delivery_partner=["A","B","C","D","E","F"]
customer_Ratings=[3,5,4,4,5,3]

df=pd.DataFrame({"Order_Time":Order_Time,"Delivery_Time":Delivery_Time,"Delivery_partner":Delivery_partner,"customer_Ratings":customer_Ratings})

df['Processing_Time'] = df['Delivery_Time'] - df['Order_Time']
print(df)

average_Time=df['Processing_Time'].mean()
print("average_Delivery_Time:",average_Time,"hours")

peak_order_Time = df['Order_Time'].mode()[0]
print(f"peak_Order_Time: {peak_order_Time}")

most_delayed_partner = df.loc[df['Processing_Time'].idxmax()]['Delivery_partner']
print(f"Most_Delayed_Partner: {most_delayed_partner}")

Top_rated_value = df['customer_Ratings'].max()
Top_rated_partners = df[df['customer_Ratings'] == Top_rated_value]['Delivery_partner'].tolist()
print(f"Top_Rated_Partner(s): {Top_rated_partners} with rating {Top_rated_value}")

df['top_rating']=(
    ((df['Processing_Time']<average_Time).astype(int))
)
print(df['top_rating'].max)

plt.bar(df['Delivery_partner'],df['Processing_Time'])
plt.xlabel('Delivery_partner')
plt.ylabel('Processing_Time')
plt.title('Delivery_Time')
plt.show()

plt.bar(df['Delivery_partner'],df['customer_Ratings'])
plt.xlabel('Delivery_partner')
plt.ylabel('customer_Ratings')
plt.title('customer_Ratings')
plt.show()
```

```python
#online shopping
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = {
    'Product purchases': ['Laptop','Mobile','Laptop','Watch','Mobile','Headphones'],
    'sales':[10000,2000,20000,1000,5000,500],
    'Cart abandonment': [2,1,0,1,1,0],
    'Customer rating': [5,2,4,3,1,5],
    'Payment methods': ['UPI','UPI','Cash','UPI','Card','Cash']
}

df = pd.DataFrame(data)

print("------ DATASET ------")
print(df)

print("-----------Most Sold Products------------")
print("Most Sold Products:",df['Product purchases'].mode().tolist())

print("---------highest revenue by customer--------")
highest_revenue=df['sales'].max()
index=df['sales'].idxmax()
product_with_high_revenue = df.loc[index, 'Product purchases']
print(f"highest revenue by customer:{highest_revenue} by {product_with_high_revenue}")


print("--------Most Used Payment Method--------")
print("Most Used Payment Method:",df['Payment methods'].mode().tolist())

print("--------Products with Poor Rating-----")
print("\nProducts with Poor Rating:")
poor = df[df['Customer rating'] < 3]
print(poor[['Product purchases','Customer rating']])

print("------charts--------")

# Most Sold Products Distribution
product_counts = df['Product purchases'].value_counts()
sns.barplot(x=product_counts.index, y=product_counts.values)
plt.title('Distribution of Product Purchases')
plt.xlabel('Product')
plt.ylabel('Number of Purchases')
plt.show()

plt.bar(df['Product purchases'],df['Cart abandonment'])
plt.xlabel('Product purchases')
plt.ylabel('Cart abandonment')
plt.title('Cart abandonment')
plt.show()

plt.bar(df['Product purchases'],df['Customer rating'])
plt.xlabel('Product purchases')
plt.ylabel('Customer rating')
plt.title('Customer rating')
plt.show()

payment_counts = df['Payment methods'].value_counts()
plt.pie(payment_counts.values, labels=payment_counts.index, autopct='%1.1f%%')
plt.title('Payment methods')
plt.show()
```

```python
#Bank Transaction
import pandas as pd

data = {
    "Transaction_Amount":[10000.00,20000.09,30000.38,40000.00,50000.90],
    "location":['city1','city2','city3','city4','city5'],
    "Time":['10:00','11:00','12:00','13:00','14:00'],
    "Account_Activity":['active','inactive','active','inactive','active']
}

df = pd.DataFrame(data)
print(df)

print("-----Unusual transaction-------")
Unusual_transaction=df[df['Transaction_Amount']!=df['Transaction_Amount'].astype(int)]
print(Unusual_transaction)

print("---------High Risk Accounts---------")
high_rish_accounts=df[df['Transaction_Amount']>25000]
print(high_rish_accounts)

print("----------Fraud patterns--------")
Fraud_patterns=df[df['Account_Activity']=='inactive']
print(Fraud_patterns)

print("---------charts------------")
plt.bar(df['location'],df['Transaction_Amount'])
plt.xlabel('location')
plt.ylabel('Transaction_Amount')
plt.title('Transaction_Amount')
plt.show()
```

```python
# SuperMarket
import pandas as pd
df = pd.DataFrame({
    "Product_name":["A","B","C","D","E"],
    "Daily_Sales": [72, 50, 45, 80, 61],
    "Product_stock": [120, 80, 150, 60, 200],
    "Expiry_Date": [50,45,60,55,44],
    "seasonal_demand": ["High", "Medium", "Low", "High", "Medium"]
})
print(df)

print("-----------Fast Moving products-------")
fast_moving_products = df[(df["Daily_Sales"] > df["Daily_Sales"].mean()) & (df["seasonal_demand"]=="High")]
print(fast_moving_products)

print("---------slow moving products-------")
slow_moving_products = df[(df["Daily_Sales"] < df["Daily_Sales"].mean()) & (df["seasonal_demand"]=="Low")]
print(slow_moving_products)

print("---------Out of Stock products-------")
out_of_stock_products = df['Product_stock']-df['Daily_Sales']*2
out_of_stock = out_of_stock_products[out_of_stock_products < 0]
print(df.loc[out_of_stock.index, 'Product_name'])

print("---------charts-------")
plt.bar(df['Product_name'],df['Daily_Sales'])
plt.xlabel('Product_name')
plt.ylabel('Daily_Sales')
plt.title('Daily_Sales')
plt.show()
```
