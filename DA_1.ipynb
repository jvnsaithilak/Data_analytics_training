```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

marks=[90,95,90,70,85,75]
hours=[7,9,8,7,9,7]

print('mean:',np.mean(marks))
print('median:',np.median(marks))
print('mode:',pd.Series(marks).mode())
print('variance:',np.var(marks))
print('standard deviation:',np.std(marks))
print('correlation:',np.corrcoef(marks,hours))

plt.figure(figsize=(5,5))
plt.hist(marks)
plt.xlabel('marks')
plt.ylabel('hours')
plt.title('marks vs hours')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('amazon.csv', encoding='latin1')

df = pd.DataFrame(data)

# Clean and convert 'discount_percentage' to numeric
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '').astype(float)

print("total no.of prodect:",len(df['product_id']))

print("-------dicount ----------")
# Now, compare numerically
discount = df[df['discount_percentage'] <= 30]
print(len(discount))
# ---------------------------------------------
# BAR CHART
# ---------------------------------------------
top_products = df.head(10)

plt.figure(figsize=(10,5))
plt.bar(top_products['product_name'], top_products['rating'])

plt.xticks(rotation=90)
plt.xlabel("Product Name")
plt.ylabel("Rating")
plt.title("Top 10 Product Ratings")

plt.tight_layout()
plt.show()

# ---------------------------------------------
# PIE CHART
# ---------------------------------------------
rating_count = df['rating'].value_counts().head(5)

plt.figure(figsize=(6,6))
plt.pie(rating_count,
        labels=rating_count.index,
        autopct='%1.1f%%')

plt.title("Rating ")
plt.show()

# ---------------------------------------------
# LINE CHART
# ---------------------------------------------
plt.figure(figsize=(10,5))

plt.plot(top_products['discount_percentage'],
         top_products['actual_price'],
         marker='o')

plt.xlabel("Discount Percentage")
plt.ylabel("Actual Price")
plt.title("Discount vs Actual Price")

plt.grid(True)
plt.show()
```

```python
import pandas as pd
tank_size=220
days=[1,2,3,4,5,6,7]
consumstion=[190,194,200,203,208,209,215]
data={'days':days,'consumstion':consumstion}
df=pd.DataFrame(data)
print(df)

avg_consumstion=df['consumstion'].mean()
print("average consumstion:",avg_consumstion)

daily_increase = df['consumstion'].diff().dropna()
print(daily_increase)
print("sum of daily water usage increase:",sum(daily_increase))
water_usage_incresed = daily_increase.mean()
print("average daily water usage increase:", water_usage_incresed)

if(((consumstion[-1])+water_usage_incresed)>tank_size):
  print("water will shortage")
else:
  print("no shortages")
```

```python
import pandas as pd
days=[1,2,3,4,5,6,7]
temp=[10,29,21,25,16,15,20]
data={'days':days,'temp':temp}
df=pd.DataFrame(data)
print(df)

avg_temp=df['temp'].mean()
print("average temperature:",avg_temp)
print(df)

daily_increase = df['temp'].diff().dropna()
print(daily_increase)
print("sum of daily temperature increase:",sum(daily_increase))
temp_increase = daily_increase.mean()
print("average daily temperature increase:", temp_increase)

if(((temp[-1])+temp_increase)>20):
  print("IF temp increase : No Rain fall")
else:
  print("IF temp increase : Rain fall")

if(((temp[-1])-temp_increase)>20):
  print("IF temp decrease : No Rain fall")
else:
  print("IF temp decrease : Rain fall")
```

```python
import pandas as pd
No_of_beds=3
name=['A','B','C','D','E']
age=[45,60,55,70,69,]
condition=['C','N','N','C','N']
oxygen_level=[90,89,95,69,59]
data={'name':name,'age':age,'condition':condition,'oxygen_level':oxygen_level}
df=pd.DataFrame(data)
print(df)


print(df[df['age'] > 60][['name','age']])
print(df[df['condition']=='C'][['name','condition']])
print(df[df['oxygen_level'] < 90][['name','oxygen_level']])

df['priority'] = (
    (df['age'] > 60).astype(int) +
    (df['condition'] == 'C').astype(int) +
    (df['oxygen_level'] < 90).astype(int)
)
df = df.sort_values(by='priority', ascending=False)
df['Bed_Status'] = 'Waiting'

for i in range(min(No_of_beds, len(df))):
    df.iloc[i, df.columns.get_loc('Bed_Status')] = 'Allocated'

print(df)
```
